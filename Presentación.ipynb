# Comenzando con R #

R es un lenguaje de programación abierto enfocado para estadistica y ciencias de datos. *R* opera en dos modos, el modo **interactivo** y modo **batch** 

Considere ahora una distribución normal estandar, recuerde que una variable aleatoria tiene una distribución normal si su función de probabilidad es:

$$ \frac{1}{2 \sigma \sqrt{2 \pi}}e^{-\frac{(x-\mu)^2}{2\sigma^2}} $$

Considere entonces $\sigma=1$ y $\mu = 0$ y $X$ está distribuida como $N(0,1)$. Es decir $X$ toma valores tanto positivos como negativos y con valores cercanos a $0$. Ahora, considere entonces una nueva variable aleatoria $Y = \left| X \right| $. Como estamos tomando el valor absoluto, los valores de $Y$ no serán centrados en $0$ y la media de $Y$ sera positiva.

Encontremos ahora la media de $Y$. **Nuestro acercamiendo se basa en un ejemplo de la simulacion de** $N(0,1)$ **varia**

```r
mean(abs(rnorm(100)))
```

```r
rnorm(7)
```

Este codigo, genera 100 *random variates*, encuentra sus valores absolutos, y por último encuentra la media de sus valores absolutos.

```r
rnorm(10)
```

# Introdución a Funciones #

En el corazón de *R* se basa completamente en funciones, un problema a solucionar podría ser el de considerar cuantos valores de mi dataset son impares.

```r
5/2
```

```r
# Contar el número de enteros impares en x
oddcount <- function(x) {
    k <- 0  #Asignar 0 a k
    for(n in x) {
        if (n %% 2 == 1) k <- k+1
    }
        return(x)
}
```

```r
oddcount(c(1,3,5))
```
